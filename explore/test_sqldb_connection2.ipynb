```
from pyprojroot import here
from sqlalchemy import create_engine
from sqlalchemy import inspect
```

```
db_path=str(here("data"))+ "/sqldb.db"
print(db_path)
db=f"sqlite:///{db_path}"
print(db)
```

```
#Create an engine that connects to SQlite Database
engine=create_engine(f"sqlite:///{db_path}")
print(engine)
```

```
#Start the connection
connection=engine.connect()
print(connection)
```

```
inspector=inspect(engine)
print(inspector)
```

```
#Retrive the table names from database
table_names=inspector.get_table_names()
print(table_names)
```

```
for table in table_names:
    print(f"Information about table:{table}")
    schema_name=inspector.get_schema_names()
    print(f"Schema name:{schema_name}")
    columns=inspector.get_columns(table)
    print(f"Column names:{columns}")
    for col in columns:
        print(f"Column Names:{col['name']}")
        print(f'Type:{col["type"]}')
        primary_keys=inspector.get_pk_constraint(table)
        foreign_keys=inspector.get_foreign_keys(table)
        print(f'Primary keys:{primary_keys}')
        print(f'Foreign keys:{foreign_keys}')

    connection.close()
```
